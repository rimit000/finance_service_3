```python
import pandas as pd

#  CSV 불러오기
df = pd.read_csv('적금_상품정보_API.csv')
```

```python
#  원본 데이터 행 수
print(f"원본 데이터 행 수: {len(df)}")
```

```python
# (금융회사명, 상품명, 저축기간, 기본금리, 최고우대금리) 기준 중복 제거
dedup_df = df.drop_duplicates(subset=['kor_co_nm', 'fin_prdt_nm', 'save_trm', 'intr_rate', 'intr_rate2'])
```

```python
# 기준으로 정렬
dedup_df = dedup_df.sort_values(['kor_co_nm', 'fin_prdt_nm', 'save_trm'])
```

```python
dedup_df
```

```python
# 필요한 컬럼만 선택
summary_df = dedup_df[['kor_co_nm', 'fin_prdt_nm', 'save_trm', 'intr_rate', 'intr_rate2']]
```

```python
#  저장
summary_df.to_csv('적금_상품_요약_정리.csv', index=False, encoding='utf-8-sig')
print(f" 요약 정리 완료: {len(summary_df)}개 /mnt/data/적금_상품_요약_정리.csv")
```
