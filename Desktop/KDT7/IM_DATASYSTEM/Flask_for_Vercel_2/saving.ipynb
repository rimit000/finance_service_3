```python
import pandas as pd

# 1. CSV 파일 불러오기
df = pd.read_csv("deposit_products.csv")
```

```python
# 2. 추출할 열과 새 이름 지정
columns_to_keep = {
    "kor_co_nm": "금융회사명",
    "fin_prdt_nm": "상품명",
    "join_way": "가입방법",
    "mtrt_int": "만기이자",
    "spcl_cnd": "우대조건",
    "join_member": "가입대상",
    "intr_rate_type_nm": "금리유형",
    "save_trm": "저축기간(개월)",
    "intr_rate": "기본금리(%)",
    "intr_rate2": "최고우대금리(%)"
}
```

```python
# 3. 필요한 열만 추출하고 컬럼명 변경
clean_df = df[list(columns_to_keep.keys())].rename(columns=columns_to_keep)
```

```python
# 4. 중복 제거
clean_df = clean_df.drop_duplicates().reset_index(drop=True)
```
